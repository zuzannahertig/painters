```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('painters.db')
c = conn.cursor()
```

## 10 most popular painters from English Wikipedia – total number of views in 2022

```python
df_group = pd.read_sql_query('SELECT article, SUM(views) AS total_views FROM painters WHERE project = "en.wikipedia" GROUP BY article ORDER BY SUM(views) DESC LIMIT 10', conn)
df_group
```

## 10 most popular painters from Polish Wikipedia – total number of views in 2022

```python
df_group = pd.read_sql_query('SELECT article, SUM(views) AS total_views FROM painters WHERE project = "pl.wikipedia" GROUP BY article ORDER BY SUM(views) DESC LIMIT 10', conn)
df_group
```

## Most viewed painters throughout the year – English Wikipedia

```python
df_group = pd.read_sql_query('SELECT article, MAX(views), timestamp FROM painters WHERE project = "en.wikipedia" GROUP BY timestamp', conn)
df_group
```

## Most viewed painters throughout the year – Polish Wikipedia

```python
df_group = pd.read_sql_query('SELECT article, MAX(views), timestamp FROM painters WHERE project = "pl.wikipedia" GROUP BY timestamp', conn)
df_group
```

## 10 most popular painters – their maximum and minimum monthly views in 2022

```python
df = pd.read_sql_query('SELECT article, MIN(views) AS min_views, MAX(views) AS max_views FROM painters GROUP BY article ORDER BY max_views DESC LIMIT 10', conn)
df
```
